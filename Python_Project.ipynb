```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataframe = pd.read_csv("/ProjectOrder.csv")
print(dataframe)
```

```python
print(dataframe['Sales'].head())
print(dataframe['Sales'].dtype)
```

def handleRate(value):
    try:
        value = str(value).strip()
        if not value or value.lower() in ['nan', 'none', '', 'null']:
            return np.nan

        value = value.split('.')[0]
        return float(value)
    except Exception:
        return np.nan
        dataframe['Sales'] = dataframe['Sales'].apply(handleRate)
        print(dataframe['Sales'].head())

```python
grouped_data = dataframe.groupby('Customer Segment')['Sales'].sum()
result = pd.DataFrame({'Sales':grouped_data})
plt.plot(result, c="orange", marker="o")
plt.xlabel("Customer Segment",c="maroon",size=12)
plt.ylabel("Sales", c="maroon", size=12)
plt.show
```

**Conclusion:** Through the line graph we can clearly state that we get our highest Sales from the Corporate Segment. Whereas, consumer and small business segment has the lowest sales. Therefore we have to comeup with ideas to attract consumers and small business.
